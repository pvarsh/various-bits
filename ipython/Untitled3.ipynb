```python
import numpy as np
import matplotlib.pyplot as plt

# for ipython only
%matplotlib inline
```

```python
x = range(10)
y = np.random.randn(10)
```

```python
fig, ax = plt.subplots()
### create the plot
ax.plot(x,y, color = 'red', marker = 'o', ls = '--') # marker style, line style 
# ax.hist(y)
# ax.scatter(x, y)

### titles, axis labels:
ax.set_title('Showing MPL')
ax.set_xlabel('Count')
ax.set_ylabel('Random')

### x and y limits
xmin, xmax = ax.get_xlim() # gets current x-min and x-max
ymin, ymax = ax.get_ylim() # gets current y-min and y-max
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

### tick marks
ax.set_xticks([0, 5, 6])
ax.set_xticklabels(['a', 'b', 'c'])

### grids
ax.grid(b=True, which='major', axis = 'x', color='gray', linestyle='--')

plt.show()
```
