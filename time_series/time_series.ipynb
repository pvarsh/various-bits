###### First order moving average process.

```python
# Xt = Z(t) + theta * Z(t-1)
n = 40
sigma_sq = 1
theta = 0.5

thetas = [0, .3, .7, 1, 1.5, 2]
plt.figure(1)
for i, theta in enumerate(thetas):
    Z = np.random.normal(loc = 0, scale = sigma_sq, size = n)
    X = Z[1:] + theta * Z[:-1]
    n_subplot = str(int(math.ceil(len(thetas)/2.0))) + str(2) + str(i)
    plt.subplot(n_subplot)
    plt.plot(X)
```
