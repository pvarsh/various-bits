```python
def readTopClasses(fips):
    ''' Reads a patent file for given fips, sorts innovations by count, computes fraction of total, returns dataframe'''
    path = "/Users/petervarshavsky/Dropbox/_NYU/Classes/Fall 2014/5001-Foundations/final paper/data/MSA_Patents"
    fileIn = "%s/1%s.csv" %(path, fips)
    # Read, sort, take top 10
    df = pd.read_csv(fileIn, usecols=['Class', 'Class Title', 'Total'], index_col = ['Class'])
    df.sort(columns = ['Total'], ascending = False, inplace = True)
    dfSum = df.Total.sum()
    df = df.iloc[0:10]
    totalColName = fips + "_total"
    fracColName = fips + "_frac"
    df.columns = ['classTitle', totalColName]
    df[fracColName] = df[totalColName]/dfSum
    return df

def classHistogram(fips):
    path = "/Users/petervarshavsky/Dropbox/_NYU/Classes/Fall 2014/5001-Foundations/final paper/data/MSA_Patents"
    fileIn = "%s/1%s.csv" %(path, fips)
    df = pd.read_csv(fileIn, usecols=['Class', 'Class Title', 'Total'], index_col = ['Class'])
    df.sort(columns = ['Total'], ascending = False, inplace = True)
    #fig, ax = plt.subplots(num = 1, figsize = (6,6))
    df.Total = np.log(df.Total)
    df = df.iloc[0:10]
    df.hist(normed = True)
    #fig.show()
    #return df
```

#### Patents into *patCounts* dataframe
The has 3 columns:
* *FIPS:* derived from file name dropping the first digit
* *period1:* sum of 2001-2005
* *period2:* sum of 2007-2011

```python
path = "/Users/petervarshavsky/Dropbox/_NYU/Classes/Fall 2014/5001-Foundations/final paper/data/MSA_Patents/"
patentFiles = glob.glob(path+"*.csv")

### read sums of patents for two periods from csv files into a list
patentCounts = []
for fIn in patentFiles:
    FIP = fIn.split('/')[-1][1:-4]
    sumFIP_1 = 0 # first 5 years
    sumFIP_2 = 0 # second 5 years
    with open(fIn, 'r') as f:
        reader = csv.reader(f)
        headerLine = reader.next()
        #print FIP, "1st 5 years:", headerLine[3:8], "1st 5 years:", headerLine[9:14]
        for line in reader:
            sumFIP_1 += sum([int(i) for i in line[2:7]])
            sumFIP_2 += sum([int(i) for i in line[9:14]])
    patentCounts.append([FIP, sumFIP_1, sumFIP_2])

### make data frame from list, name columns, set index
patCounts = pd.DataFrame(patentCounts)
patCounts.columns = ["FIPS", "period1pat", "period2pat"]
patCounts.index = patCounts.FIPS
patCounts.drop('FIPS', 1, inplace = True)
patCounts.head(2)
```

### Problem 1 master data frame

```python
# patCounts, gdpMain, popMain
p1df = pd.merge(gdpMain[['MSAname', 'period1gdp', 'period2gdp']],\
                      popMain[['MSAname', 'period1pop', 'period2pop']], how = 'outer', left_index = True, right_index = True)
p1df = pd.merge(p1df, patCounts,\
                      how = 'outer', left_index = True, right_index = True)
p1df.head(3)
### saving MSAs with null values
nanMSAs = p1df[p1df.MSAname_x.isnull() | p1df.MSAname_y.isnull() | p1df.period1pat.isnull()]
### MSAs without null values
p1df.dropna(how = 'any', inplace = True)
p1df.drop('MSAname_y', 1, inplace = True)
p1df.head(2)
```

```python
p1df['patIntensity1'] = (p1df.period1pat * 100000.) / p1df.period1pop
p1df['patIntensity2'] = (p1df.period2pat * 100000.) / p1df.period2pop
p1df['patIntensityChange'] = (p1df.patIntensity2 - p1df.patIntensity1)/p1df.patIntensity1
p1df.period2gdp
p1df.head(3)
```

```python
y, X = patsy.dmatrices('np.log(period1pat) ~ np.log(period1pop)', data = p1df)
modWest = sm.OLS(y, X).fit()
slopeWest = modWest.params[1]
print modWest.conf_int()

fig11, ax11 = plt.subplots(figsize = (8,6))
ax11.plot(np.log(p1df.period1pop), np.log(p1df.period1pat), 'o', label = 'MSA')
ax11.plot(np.log(p1df.period1pop), modWest.predict(), label = r'$\hat{\beta_1} = %2.3f$' %slopeWest)
ax11.plot([],[], label = r'$R^2 = %.2f$' %modWest.rsquared, alpha = 0)
ax11.legend(loc = 2)
ax11.set_title('Population and patent production')
ax11.set_xlabel('Avg population 2001-2005 (log)')
ax11.set_ylabel('Total patent production 2001-2005 (log)')
fig11.show()
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

data_dist = pdist(patentMSAbinary) # computing the distance
data_link = linkage(data_dist) # computing the linkage

dendrogram(data_link,labels = patentMSAbinary.index)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.suptitle('Samples clustering', fontweight='bold', fontsize=14);
```

```python
data = np.array(patentMSAbinary)
# Compute and plot first dendrogram.
fig = plt.figure(figsize=(8,8))
# x ywidth height
ax1 = fig.add_axes([0.05,0.1,0.2,0.6])
Y = linkage(data_dist, method='single')
Z1 = dendrogram(Y, orientation='right',labels=data.dtype.names) # adding/removing the axes
ax1.set_xticks([])

# Compute and plot second dendrogram.
ax2 = fig.add_axes([0.3,0.71,0.6,0.2])
Z2 = dendrogram(Y)
ax2.set_xticks([])
ax2.set_yticks([])

#Compute and plot the heatmap
axmatrix = fig.add_axes([0.3,0.1,0.6,0.6])
idx1 = Z1['leaves']
idx2 = Z2['leaves']
D = squareform(data_dist)
D = D[idx1,:]
D = D[:,idx2]
im = axmatrix.matshow(D, aspect='auto', origin='lower', cmap=plt.cm.YlGnBu)
axmatrix.set_xticks([])
axmatrix.set_yticks([])

# Plot colorbar.
axcolor = fig.add_axes([0.91,0.1,0.02,0.6])
plt.colorbar(im, cax=axcolor)
```
