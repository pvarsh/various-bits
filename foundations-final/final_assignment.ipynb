```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import csv
import patsy
import statsmodels.api as sm

%matplotlib inline
plt.style.use('ggplot')
```

# Read data

##### Define lists for 5-year periods

```python
period1list = ['2001', '2002', '2003', '2004', '2005']
period2list = ['2007', '2008', '2009', '2010', '2011']
```

#### Population data into *popMain* dataframe

```python
dataPath = "/Users/petervarshavsky/Dropbox/_NYU/Classes/Fall 2014/5001-Foundations/final paper/data"
fname = "population-msa-2001-2011.csv"
fileIn = dataPath + "/" + fname
popMain = pd.read_csv(fileIn, header = 4)
popMain.columns = [u'FIPS', u'MSAname', u'2001', u'2002', u'2003', u'2004', u'2005', u'2006', u'2007', u'2008', u'2009', u'2010', u'2011']
popMain.index = popMain.FIPS
popMain.drop('FIPS', 1, inplace = True)

### drop NaNs (footnotes)
popMain = popMain[popMain.MSAname.notnull()]

### Compute average GDP for two periods
popMain['period1pop'] = popMain[period1list].apply(sum, 1)/5.0
popMain['period2pop'] = popMain[period2list].apply(sum, 1)/5.0
popMain.head(2)
```

#### GDP data into *gdpMain* dataframe

```python
dataPath = "/Users/petervarshavsky/Dropbox/_NYU/Classes/Fall 2014/5001-Foundations/final paper/data"
fname = "gdp-msa-2001-2011.csv"
fileIn = dataPath + "/" + fname
gdpMain = pd.read_csv(fileIn, header = 4)
gdpMain.columns = [u'FIPS', u'MSAname', u'2001', u'2002', u'2003', u'2004', u'2005', u'2006', u'2007', u'2008', u'2009', u'2010', u'2011']
gdpMain.index = gdpMain.FIPS
gdpMain.drop('FIPS', 1, inplace = True)

### Drop NaN (footnotes)
gpdMain = gdpMain[gdpMain.MSAname.notnull()]

### Compute average GDP for two periods
gdpMain['period1gdp'] = gdpMain[period1list].apply(sum, 1)/5.0
gdpMain['period2gdp'] = gdpMain[period2list].apply(sum, 1)/5.0
gdpMain.head(2)
```

#### Compute GPD averages for two periods

#### Patents into *patCounts* dataframe
The has 3 columns:
* *FIPS:* derived from file name dropping the first digit
* *period1:* sum of 2001-2005
* *period2:* sum of 2007-2011

```python
path = "/Users/petervarshavsky/Dropbox/_NYU/Classes/Fall 2014/5001-Foundations/final paper/data/MSA_Patents/"
patentFiles = glob.glob(path+"*.csv")

### read sums of patents for two periods from csv files into a list
patentCounts = []
for fIn in patentFiles:
    FIP = fIn.split('/')[-1][1:-4]
    sumFIP_1 = 0 # first 5 years
    sumFIP_2 = 0 # second 5 years
    with open(fIn, 'r') as f:
        reader = csv.reader(f)
        headerLine = reader.next()
        #print FIP, "1st 5 years:", headerLine[3:8], "1st 5 years:", headerLine[9:14]
        for line in reader:
            sumFIP_1 += sum([int(i) for i in line[2:7]])
            sumFIP_2 += sum([int(i) for i in line[9:14]])
    patentCounts.append([FIP, sumFIP_1, sumFIP_2])

### make data frame from list, name columns, set index
patCounts = pd.DataFrame(patentCounts)
patCounts.columns = ["FIPS", "period1pat", "period2pat"]
patCounts.index = patCounts.FIPS
patCounts.drop('FIPS', 1, inplace = True)
patCounts.head(2)
```

### Problem 1 master data frame

```python
# patCounts, gdpMain, popMain
p1df = pd.merge(gdpMain[['MSAname', 'period1gdp', 'period2gdp']],\
                      popMain[['MSAname', 'period1pop', 'period2pop']], how = 'outer', left_index = True, right_index = True)
p1df = pd.merge(p1df, patCounts,\
                      how = 'outer', left_index = True, right_index = True)
p1df.head(3)
### saving MSAs with null values
nanMSAs = p1df[p1df.MSAname_x.isnull() | p1df.MSAname_y.isnull() | p1df.period1pat.isnull()]
### MSAs without null values
p1df.dropna(how = 'any', inplace = True)
p1df.head(2)
```

```python
p1df['patIntensity'] = (p1df.period1pat * 100000.) / p1df.period1pop
p1df['gdpCapita'] = p1df.period2gdp# / p1df.period2pop
p1df.head(3)
```

```python
fig0, ax0 = plt.subplots(figsize = (10, 5))
ax0.plot(np.log(p1df.patIntensity), np.log(p1df.gdpCapita), marker = 'o', lw = 0)
plt.show()
```

```python
y, X = patsy.dmatrices('np.log(gdpCapita) ~ np.log(patIntensity)', data = p1df)
mod1 = sm.OLS(y, X).fit()
mod1.summary()
```
