```python
import numpy as np
import scipy as sp
import scipy.ndimage as nd
from matplotlib import pyplot as plt

%matplotlib inline

x = nd.imread("mona-lisa-rgb.jpg")
x.shape
```

```python
type(x[0,0,0])
```

```python
height = x.shape[0]
width = x.shape[1]

fig1, ax1 = plt.subplots(1,1, figsize = (10, 6))
#fig6.subplots_adjust(0, 0, 1, 1, 0, 0)
#[i.axis('off') for i in ax6]
im1 = ax1.imshow(x)
fig1.canvas.draw()
plt.show()
```

```python
height = x.shape[0]
width = x.shape[1]

fig2, ax2 = plt.subplots(1,1, figsize = (10, 6))
#fig6.subplots_adjust(0, 0, 1, 1, 0, 0)
#[i.axis('off') for i in ax6]
im2 = ax2.imshow(x[0:height/2, 0:width/2])
fig2.canvas.draw()
plt.show()
```

```python
height = x.shape[0]
width = x.shape[1]

fig3, ax3 = plt.subplots(1,1, figsize = (10, 6))
#fig6.subplots_adjust(0, 0, 1, 1, 0, 0)
#[i.axis('off') for i in ax6]
im3 = ax3.imshow(x[height/2:, width/2:])
fig3.canvas.draw()
plt.show()
```

```python
height = x.shape[0]
width = x.shape[1]
centrHalf = (1-np.sqrt(.5))/2
fig4, ax4 = plt.subplots(1,1, figsize = (10, 6))
#fig6.subplots_adjust(0, 0, 1, 1, 0, 0)
#[i.axis('off') for i in ax6]
im4 = ax4.imshow(x[(centrHalf*height):(height - centrHalf*height), (centrHalf*width):(width - centrHalf*width)])
fig4.canvas.draw()
plt.show()
```
